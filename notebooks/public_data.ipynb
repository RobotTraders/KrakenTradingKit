# Public Data API Interaction Example

## Setup

```python
from kraken_kit import SpotConnector, FuturesConnector

spot = SpotConnector.public()
futures = FuturesConnector.public()
```

## Futures Instrument Info

```python
futures.get_instruments()
```

## Futures Ticker

```python
futures_symbol = "PF_XBTUSD"
ticker = futures.get_ticker(futures_symbol)
ticker
```

## Futures Order Book

```python
ob = futures.get_order_book(futures_symbol)
ob
```

```python
ob['asks']
```

```python
best_ask = ob["asks"][0][0]
best_bid = ob["bids"][0][0]
best_ask, best_bid
```

## Futures Spread

```python
futures.get_spreads(futures_symbol)
```

## Futures Recent Trades

```python
count = 10

futures.get_recent_trades(futures_symbol, count=count)
```

## Futures OHLCV

```python
timeframe = "1d"

ohlcv = futures.get_ohlcv(futures_symbol, timeframe=timeframe)
ohlcv 
```

```python
timeframe = "1h"
start_date = "2026-04-01"

ohlcv = futures.get_ohlcv(futures_symbol, timeframe=timeframe, start_date=start_date)
ohlcv 
```

```python
timeframe = "1h"
start_date = "2026-01-01"
end_date = "2026-03-01"

ohlcv = futures.get_ohlcv(futures_symbol, timeframe=timeframe, start_date=start_date, end_date=end_date)
ohlcv 
```

```python
timeframe = "1h"
start_date = "2025-01-01"
end_date = "2025-07-01"

ohlcv = futures.get_ohlcv(futures_symbol, timeframe=timeframe, start_date=start_date, end_date=end_date)
ohlcv 
```

```python
ohlcv.to_csv('btc_usdc.csv')
```

```python
import pandas as pd

ohlcv = pd.read_csv('btc_usdc.csv', index_col='time', parse_dates=True)
ohlcv
```
